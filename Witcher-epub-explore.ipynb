```python
import ebooklib
from ebooklib import epub
from ebooklib.utils import debug
from bs4 import BeautifulSoup
```

```python
book = epub.read_epub('books/Baptism of Fire.epub')
```

```python
book.get_metadata('DC', 'title')[0][0]
```

```python
book.get_metadata('DC', 'date')[0][0]
```

```python
book.spine
```

Only need consecutive chapters

```python
spine_list = [x[0] for x in book.spine]
```

```python
chapter_list = []
for i, x in enumerate(spine_list):
    if 'chapter' in x or 'epilogue' in x:
        if chapter_list:
            if spine_list.index(chapter_list[-1]) + 1 == i:
                chapter_list.append(x)
        else:
            chapter_list.append(x)
```

```python
chapter_list
```

```python
chapters = [
    BeautifulSoup(book.get_item_with_id(chapter).get_content(), 'lxml')
    for chapter in chapter_list
]
```

```python
[bool(chapter.select('[class*="dl"]')) for chapter in chapters]
```
